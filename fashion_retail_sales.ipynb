```python
#total sales over time using line plot
sales_time = fashion_sales_update_dataset.groupby(fashion_sales_update_dataset['Date Purchase'].dt.to_period("M"))['Purchase Amount (USD)'].sum()
sales_time.index = sales_time.index.to_timestamp()


plt.figure(figsize=(10,6))
sales_time.plot(marker='o')# line plot
plt.title("Total Sales Over Time")
plt.ylabel("Total Purchase Amount")
plt.xlabel("Date")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#top perchased items using bar plot
top_items = fashion_sales_update_dataset.groupby('Item Purchased')['Purchase Amount (USD)'].sum().sort_values(ascending=False).head(10)


plt.figure(figsize=(10,6))
sns.barplot(x=top_items.values, y=top_items.index)
plt.title("Top 10 Purchased Items by Revenue")
plt.xlabel("Total Revenue")
plt.ylabel("Item Purchased")
plt.tight_layout()
plt.show()
```

```python
#purchase amount distribution using histogram and box plot
plt.figure(figsize=(10,5))
sns.histplot(fashion_sales_update_dataset['Purchase Amount (USD)'], bins=30, kde=True)
plt.title("Purchase Amount Distribution")
plt.xlabel("Purchase Amount")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,4))
sns.boxplot(x=fashion_sales_update_dataset['Purchase Amount (USD)'], color="skyblue")
plt.title("Boxplot - Purchase Amount")
plt.tight_layout()
plt.show()
```

```python
#rating by product using boxplot
avg_rating = fashion_sales_update_dataset.groupby('Item Purchased')['Review Rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_rating.values, y=avg_rating.index)
plt.title("Top 10 Products by Average Review Rating")
plt.xlabel("Average Rating")
plt.ylabel("Item Purchased")
plt.tight_layout()
plt.show()
```

```python
#rating vs purchase amount using scatter plot
plt.figure(figsize=(10,6))
sns.scatterplot(data=fashion_sales_update_dataset, x='Purchase Amount (USD)', y='Review Rating', alpha=0.6)
plt.title("Review Rating vs Purchase Amount")
plt.xlabel("Purchase Amount")
plt.ylabel("Review Rating")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#payment mrthod popularity
payment_counts = fashion_sales_update_dataset['Payment Method'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=payment_counts.index, y=payment_counts.values)
plt.title("Popularity of Payment Methods")
plt.xlabel("Payment Method")
plt.ylabel("Number of Transactions")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


plt.figure(figsize=(6,6))
plt.pie(payment_counts, labels=payment_counts.index, autopct="%1.1f%%", startangle=140, colors=sns.color_palette("Set2"))
plt.title("Payment Method Share")
plt.tight_layout()
plt.show()
```

```python
#monthly transactions count
monthly_txn = fashion_sales_update_dataset['Date Purchase'].dt.to_period("M").value_counts().sort_index()
monthly_txn.index = monthly_txn.index.to_timestamp()

plt.figure(figsize=(10,5))
monthly_txn.plot(marker='o')
plt.title("Monthly Count of Transactions")
plt.xlabel("Date")
plt.ylabel("Number of Transactions")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Ensure you have an independent DataFrame
fashion_sales_update_dataset = fashion_sales_update_dataset.copy()

# Create a 'Month' column
fashion_sales_update_dataset['Month'] = fashion_sales_update_dataset['Date Purchase'].dt.to_period('M')

# Pivot to create a matrix of customer spend (rows: customers, columns: month)
customer_monthly = fashion_sales_update_dataset.pivot_table(
    index='Customer Reference ID', 
    columns='Month', 
    values='Purchase Amount (USD)', 
    aggfunc='sum'
).fillna(0)

# Plot the heatmap
plt.figure(figsize=(12,8))
sns.heatmap(customer_monthly, cmap="YlGnBu", linewidths=0.1)
plt.title("Customer Spending Patterns by Month")
plt.xlabel("Month")
plt.ylabel("Customer ID")
plt.tight_layout()
plt.show()
```
